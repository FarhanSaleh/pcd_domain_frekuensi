```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk menampilkan gambar
def show_image(image, title="Image"):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Membaca gambar
image = cv2.imread('foto.jpg')

# Convert gambar ke grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Difference Image
diff_image = cv2.absdiff(image, cv2.bitwise_not(image))

# Average Filter
avg_filter = cv2.blur(image, (5, 5))

# Median Filter
median_filter = cv2.medianBlur(image, 5)

# Min Filter
min_filter = cv2.erode(image, np.ones((5,5),np.uint8))

# Max Filter
max_filter = cv2.dilate(image, np.ones((5,5),np.uint8))

# Laplacian
laplacian = cv2.Laplacian(gray_image, cv2.CV_64F)

# Gradient Sobel
gradient_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)
gradient_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=5)
gradient_sobel = cv2.magnitude(gradient_x, gradient_y)

```

```python
# Menampilkan gambar
show_image(diff_image, "Difference Image")
show_image(avg_filter, "Average Filter")
show_image(median_filter, "Median Filter")
show_image(min_filter, "Min Filter")
show_image(max_filter, "Max Filter")
show_image(np.uint8(np.absolute(laplacian)), "Laplacian")
show_image(np.uint8(gradient_sobel), "Gradient Sobel")
```
